```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
%matplotlib inline

import sklearn
```

```python
data = pd.read_csv('data_features.csv')
data.shape
```

```python
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
data.head()
```

```python
data['lag'] = data['buy_time'] - data['feat_time']
data.head()
```

```python
#Для записей в которых дата продажи раньше даты признаков устанавливаем максимальный лаг+отклонение
max_lag = data['lag'].max()
msk = data['lag'] < 0
data.loc[msk, 'lag'] = max_lag - data['lag']
```

```python
#Берем модуль от лага
#data['lag'] = np.abs(data['lag'])
```

```python
data.head()
```

```python
grp = data.groupby('offer_id').agg({'lag': min}).reset_index()
grp
```

```python
data = grp.merge(data, how='left', on=['offer_id', 'lag'])
```

```python
data.head()
```

```python
data.shape
```

```python
msk = data['vas_id'] == 1
data_vas1 = data.loc[msk, :]
data_vas1.head()
```

```python
for i in [1,2,4,5,6,7,8,9]:
    msk = data['vas_id'] == i
    data_vas = data.loc[msk, :]
    data_vas.to_csv('data_vas'+str(i)+'.csv')
```
